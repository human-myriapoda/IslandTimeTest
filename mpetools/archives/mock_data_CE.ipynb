```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, ifft
```

# Mock data

```python
# Seed
np.random.seed(42)

# Date range
date_range = np.arange(2000, 2016, 1)

# Coastal erosion
coastal_erosion = -0.3 * np.arange(len(date_range)) + 0.8*np.random.normal(size=len(date_range)) + 15

# Sea-level rise
sea_level = 0.05 * np.arange(len(date_range)) + 0.8*np.random.normal(size=len(date_range)) + 2

# Construction of seawall / land reclamation
v = np.linspace(-1, 1, len(date_range))
sea_wall = 2. * np.heaviside(v, 0.5) + 3. + 0.1 * np.random.normal(size=len(v))
#plt.plot(range(len(out)), out)

plt.plot(date_range, coastal_erosion, label='CE')
plt.plot(date_range, sea_level, label='SL')
plt.plot(date_range, sea_wall, label='SW')
#plt.plot(date_range, np.convolve(sea_wall, coastal_erosion))
plt.legend()
```
