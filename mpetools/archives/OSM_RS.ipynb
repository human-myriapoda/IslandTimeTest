```python
def affine_function(x, m, b):
    return m*x + b

for bui in range(0, 2):#range(len(buildings['geometry'])):
    buildings_arr = np.array(buildings['geometry'][bui].exterior.coords)
    buildings_arr_fill = 0.
    plt.plot(buildings_arr[:, 0], buildings_arr[:, 1])

    m = []
    b = []

    # Affine functions (lines)
    for nl in range(buildings_arr.shape[0]-1):
        #print(buildings_arr[nl, :], buildings_arr[nl+1, :])
        x1 = buildings_arr[nl, 0]
        x2 = buildings_arr[nl+1, 0]
        y1 = buildings_arr[nl, 1]
        y2 = buildings_arr[nl+1, 1]
        lon_range = np.linspace(x1, x2, 30)
        m.append((y2-y1)/(x2-x1))
        b.append(y1 - m[nl]*x1)
        #plt.plot(lon_range, m[nl]*(lon_range) + b[nl])

    x_range_t = np.linspace(np.min(buildings_arr[:, 0]), np.max(buildings_arr[:, 0]), 30)
    y_range_t = np.linspace(np.min(buildings_arr[:, 1]), np.max(buildings_arr[:, 1]), 30)

    x_full = []
    y_full = []

    for xx in x_range_t[1:-1]:
        boundaries = np.sort(np.unique(np.array([np.round(affine_function(xx, m[i], b[i]), 7) for i in range(buildings_arr.shape[0]-1)])))
        #print(boundaries)
        boundaries_middle = boundaries[int(len(boundaries)/2)-1:-1*(int(len(boundaries)/2-1))]
        #print('bm', boundaries_middle)
        for yy in y_range_t[1:-1]:
            if yy < boundaries_middle[1] and yy > boundaries_middle[0]:
                x_full.append(xx)
                y_full.append(yy)
                plt.scatter(xx, yy)

    buildings_arr_fill = np.concatenate((buildings_arr, np.column_stack((np.array(x_full), np.array(y_full)))))

#for i in range(len(x_range_t)):
    #plt.axvline(x_range_t[i])
    #plt.axhline(y_range_t[i])
```

```python
for bui in range(len(buildings['geometry'])):
    buildings_poly = buildings['geometry'][bui]

    buildings_arr = np.array(buildings_poly.exterior.coords)

    #plt.plot(buildings_arr[:, 0], buildings_arr[:, 1])

    x_range_t, y_range_t = np.meshgrid(np.linspace(np.min(buildings_arr[:, 0]), np.max(buildings_arr[:, 0]), 10), \
                                       np.linspace(np.min(buildings_arr[:, 1]), np.max(buildings_arr[:, 1]), 10))
    
    x_range_t = np.ravel(x_range_t)
    y_range_t = np.ravel(y_range_t)

    arr_points = np.array([Point(x, y) for (x, y) in zip(x_range_t, y_range_t)])

    arr_cond = buildings_poly.contains(arr_points)

    arr_points_cond = np.column_stack((x_range_t[np.argwhere(arr_cond)].reshape((-1, )), y_range_t[np.argwhere(arr_cond)].reshape((-1, ))))

    #plt.scatter(arr_points_cond[:, 0], arr_points_cond[:, 1])

    if bui == 0:
        buildings_arr_fill = np.concatenate((buildings_arr, arr_points_cond))
    
    else:
        buildings_arr_fill = np.concatenate((buildings_arr_fill, np.concatenate((buildings_arr, arr_points_cond))))
```
